```python
df['Season'] = df['Season'].astype(str)
df.loc[(df['Season'].str[-2:] == '21') & (df['Season'].str.len() == 6),'Season'] = 'Autumn ' + df.loc[df['Season'].str[-2:] == '21','Season'].str[:4]
df.loc[(df['Season'].str[-2:] == '22') & (df['Season'].str.len() == 6),'Season'] = 'Holiday ' + df.loc[df['Season'].str[-2:] == '22','Season'].str[:4]
df.loc[(df['Season'].str[-2:] == '11') & (df['Season'].str.len() == 6),'Season'] = 'Spring ' + df.loc[df['Season'].str[-2:] == '11','Season'].str[:4]
df.loc[(df['Season'].str[-2:] == '12') & (df['Season'].str.len() == 6),'Season'] = 'Summer ' + df.loc[df['Season'].str[-2:] == '12','Season'].str[:4]
```

```python
# Add Leading Zero to color
df['Color'] = df['Color'].astype(str)
df['Color'] = df['Color'].apply(lambda x: x.zfill(2))
```

```python
# Key Generation
df['StyleSupplier'] = df['Style'].astype(str) + df['Supplier code']
df['ArticleSupplier'] = df['Style'].astype(str) + df['Color'].astype(str) + df['Supplier code']
df['ArticleSeasonSourcingOrganization'] = df['Style'].astype(str) + ' ' + df['Color'].astype(str) + df['Season'] + df['SourcingOrganization']
```

```python
# Calculate Style Total

styleTotal = df.loc[:,['StyleSupplier','Order Quantity']].groupby('StyleSupplier').sum()
styleTotal = styleTotal.rename(columns={'Order Quantity': 'Style Total'})
df = pd.merge(df, styleTotal, how='left', on=['StyleSupplier'])
```

```python
# Calculate Article Total

articleTotal = df.loc[:,['ArticleSupplier','Order Quantity']].groupby('ArticleSupplier').sum()
articleTotal = articleTotal.rename(columns={'Order Quantity': 'Article Total'})
df = pd.merge(df, articleTotal, how='left', on=['ArticleSupplier'])
```

```python
# OPD check

df.loc[(df['SourcingOrganization'] == 'WCA') | (df['SourcingOrganization'] =='WCM'),'OPDFormula'] ='OK'
df.loc[(df['Product Character']=='PROMOTION') | (df['Order reason']=='PRO'),'OPDFormula'] ='OK'
df.loc[(df['Product Character'] =='Inline Product') & (df['POD'].str[:7] == 'National') & (df['Development Type'].str[:8] == 'Resourced'), 'OPDFormula'] ='OK'
df.loc[df['Product Character'] != 'Inline Product','OPDFormula'] ='OK'
df.loc[df['ArticleSeasonSourcingOrganization'].isin(productLine['ArticleSeasonSourcingOrganization']), 'OPDFormula'] ='OK'
df.loc[df['OPDFormula'].isnull(),'OPDFormula'] ='Wrong OPD'
df['OPDFormula'].isnull().sum()
```
